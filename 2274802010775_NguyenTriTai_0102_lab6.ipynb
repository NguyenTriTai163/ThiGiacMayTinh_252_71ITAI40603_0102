```python
import cv2
import matplotlib.pyplot as Tai

# Đọc ảnh và đổi màu
img = cv2.imread('image.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Load file XML
stop_data = cv2.CascadeClassifier('stop_data.xml')

# Kiểm tra xem bộ phân loại có được tải thành công không
if stop_data.empty():
    print("Lỗi: Không thể tải tệp 'stop_data.xml'. Vui lòng đảm bảo tệp này tồn tại.")
else:
    # Nhận diện
    found = stop_data.detectMultiScale(img_gray, minSize =(20, 20))
    amount_found = len(found)

    if amount_found != 0:
        for (x, y, width, height) in found:
            cv2.rectangle(img_rgb, (x, y), (x + height, y + width), (0, 255, 0), 5)

    # Hiển thị ảnh
    Tai.imshow(img_rgb)
    Tai.show()
```

```python
import cv2
import matplotlib.pyplot as Tai

# Sử dụng ảnh hinh2.jpg
img2 = cv2.imread('image2.jpg')
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Load file XML nhận diện khuôn mặt
face_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Nhận diện
faces = face_data.detectMultiScale(img_gray2, minSize=(30, 30))

if len(faces) != 0:
    for (x, y, width, height) in faces:
        cv2.rectangle(img_rgb2, (x, y), (x + width, y + height), (255, 0, 0), 5) # Viền màu xanh dương

# Hiển thị ảnh
Tai.imshow(img_rgb2)
```

```python

# Khuyến khích dùng lại hinh2.jpg (ảnh có khuôn mặt) hoặc dùng hinh3.jpg
img3 = cv2.imread('image3.jpg')
img_gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
img_rgb3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

# Load file XML nhận diện mắt
eye_data = cv2.CascadeClassifier('haarcascade_eye.xml')

# Nhận diện (thường mắt nhỏ hơn mặt nên để minSize nhỏ lại)
eyes = eye_data.detectMultiScale(img_gray3, minSize=(10, 10))

if len(eyes) != 0:
    for (x, y, width, height) in eyes:
        cv2.rectangle(img_rgb3, (x, y), (x + width, y + height), (0, 0, 255), 3) # Viền màu đỏ

# Hiển thị ảnh
Tai.imshow(img_rgb3)
```

```python


img4 = cv2.imread('image4.jpg')
img_gray4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
img_rgb4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

# Load file XML nhận diện nụ cười
smile_data = cv2.CascadeClassifier('haarcascade_smile.xml')

# Nhận diện nụ cười (chỉnh scaleFactor và minNeighbors để tránh nhận diện nhầm)
smiles = smile_data.detectMultiScale(img_gray4, scaleFactor=1.8, minNeighbors=20, minSize=(25, 25))

if len(smiles) != 0:
    for (x, y, width, height) in smiles:
        cv2.rectangle(img_rgb4, (x, y), (x + width, y + height), (255, 255, 0), 3) # Viền màu vàng

# Hiển thị ảnh
Tai.imshow(img_rgb4)
```
