```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as Tai

img = cv.imread('hinh1.jpg', 0)
```

```python
ret, thresh1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)

Tai.imshow(thresh1, cmap='gray')
Tai.title('Thresholding')
Tai.show()
```

```python
ret2, thresh_otsu = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

Tai.imshow(thresh_otsu, cmap='gray')
Tai.title('Otsu Algorithm')
Tai.show()
```

```python
img_color = cv.imread('hinh1.jpg')
img_rgb = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)
Z = img_rgb.reshape((-1,3))
Z = np.float32(Z)

criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 4
ret, label, center = cv.kmeans(Z, K, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)

center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((img_rgb.shape))

Tai.imshow(res2)
Tai.title('K-means Clustering')
Tai.show()
```

```python
img_floodfill = img.copy()
h, w = img.shape[:2]
mask = np.zeros((h+2, w+2), np.uint8)

cv.floodFill(img_floodfill, mask, (0,0), 255)

Tai.imshow(img_floodfill, cmap='gray')
Tai.title('Region Growing')
Tai.show()
```

```python
edges = cv.Canny(img, 100, 200)

Tai.imshow(edges, cmap='gray')
Tai.title('Edge-based Segmentation')
Tai.show()
```
